```python
import pandas as pd
import PyPDF2 as Pypdf
import re
```

```python
# Adapted from https://stackoverflow.com/questions/2481945/how-to-read-line-by-line-in-pdf-file-using-pypdf
def getPDFContent(path, first_page, last_page):
    content = ""
    pdf = Pypdf.PdfFileReader(path)
    for i in range(first_page, last_page):
        content += pdf.getPage(i).extractText() + "\n"
    content = " ".join(content.replace(u"\xa0", " ").strip().split())
    return content
```

```python
text = getPDFContent('GSS_Codebook.pdf', 125, 226)
text
```

```python
question_tuples = re.findall('(\s[A-Z0-9]+\.\s)', text)
question_tuples[0:5]
```
